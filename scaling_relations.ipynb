```python
import observation as ob
```

```python
me = ob.observation().classical().new_RGB().new_MS().new_fdnu()
```

```python
print('''
-----------------------------------------------------------------------
                  scaling relations: classical
-----------------------------------------------------------------------
ID                  mass              radius            logg
-----------------------------------------------------------------------''')
for i in range(len(me.id)):
    print('{:15s}'.format('TIC '+me.id[i]),'   ',me.classical_mass[i], '   ',me.classical_radius[i],'   ', me.classical_logg[i])
```

```python
print('''
-----------------------------------------------------------------------
                  scaling relations: new_RGB
-----------------------------------------------------------------------
ID                  age               mass              radius
-----------------------------------------------------------------------''')
for i in range(len(me.id)):
    print('{:15s}'.format('TIC '+me.id[i]),'   ',me.new_RGB_age[i],'   ',me.new_RGB_mass[i], '   ',me.new_RGB_radius[i])
```

```python
print('''
-----------------------------------------------------------------------
                  scaling relations: new_MS
-----------------------------------------------------------------------
ID                  age               mass              radius
-----------------------------------------------------------------------''')
for i in range(len(me.id)):
    print('{:15s}'.format('TIC '+me.id[i]),'   ',me.new_MS_age[i],'   ',me.new_MS_mass[i], '   ',me.new_MS_radius[i])
```

```python
print('''
-----------------------------------------------------------------------
                  scaling relations: fdnu
-----------------------------------------------------------------------
ID                  mass              radius            fdnu
-----------------------------------------------------------------------''')
for i in range(len(me.id)):
    print('{:15s}'.format('TIC '+me.id[i]),'   ',me.fdnu_mass[i], '   ',me.fdnu_radius[i],'   ','%.3f'%me.fdnu[i])
```
